But this playground competition's dataset proves that much more influences price negotiations than the number of bedrooms or a white-picket fence.

With 79 explanatory variables describing (almost) every aspect of residential homes in Ames, Iowa, this competition challenges you to predict the final price of each home.

# Updating Missing Values in Train Data

```python
# From feature description we can see that 
# NA in Alley means "No Access to Alley". Similary,
# NA in FireplaceQu means "No Fireplace",
# NA in PoolQC means "No Pool"
# NA in Fence means "No Fence"
# NA in MiscFeature means "No MiscFeature"
# NA in GarageQual,GarageCond,GarageType and GarageFinish means "No Garage"
# NA in GarageYrBlt is updated with YearBuilt as for most of the data points YearBuilt and GarageYrBlt is same as seen above.
# NA in BsmtQual,BsmtCond,BsmtExposure,BsmtFinType1,BsmtFinType2 means "No Basement"

train['Alley']=train.Alley.apply(lambda x: 'No Alley' if pd.isnull(x) else x)
train['FireplaceQu']=train.FireplaceQu.apply(lambda x: 'No Fireplace' if pd.isnull(x) else x)
train['PoolQC']=train.PoolQC.apply(lambda x: 'No Pool' if pd.isnull(x) else x)
train['Fence']=train.Fence.apply(lambda x: 'No Fence' if pd.isnull(x) else x)
train['MiscFeature']=train.MiscFeature.apply(lambda x: 'No MiscFeature' if pd.isnull(x) else x)
train['GarageQual']=train.GarageQual.apply(lambda x: 'No Garage' if pd.isnull(x) else x)
train['GarageCond']=train.GarageCond.apply(lambda x: 'No Garage' if pd.isnull(x) else x)
train['GarageType']=train.GarageType.apply(lambda x: 'No Garage' if pd.isnull(x) else x)
train['GarageFinish']=train.GarageFinish.apply(lambda x: 'No Garage' if pd.isnull(x) else x)
train['GarageYrBlt']=train.GarageYrBlt.apply(lambda x:0 if pd.isnull(x) else x)
train['BsmtQual']=train.BsmtQual.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
train['BsmtCond']=train.BsmtCond.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
train['BsmtFinType1']=train.BsmtFinType1.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
train['BsmtFinType2']=train.BsmtFinType2.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
```

```python
# From above observation if BsmtQual,BsmtCond,BsmtExposure,BsmtFinType1,BsmtFinType2 all have NA then its "No Basement" else if 
# BsmtQual is NA but others are not then NA is replaced with "TA". Similarly, BsmtCond is NA but others are not then NA is replaced with "TA",
# BsmtExposure is NA but others are not then NA is replaced with "No".

test['BsmtQual']=test[['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1',
                         'BsmtFinType2']].apply(lambda x: 'No Basement' if pd.isnull(x['BsmtQual']) and pd.isnull(x['BsmtCond']) and
                                                pd.isnull(x['BsmtExposure']) and pd.isnull(x['BsmtFinType1']) and 
                                                pd.isnull(x['BsmtFinType2']) else 'TA' if pd.isnull(x['BsmtQual']) and (np.all(pd.notnull(x['BsmtCond'])))
                                                and np.all((pd.notnull(x['BsmtExposure']))) and np.all((pd.notnull(x['BsmtFinType1']))) and 
                                                np.all((pd.notnull(x['BsmtFinType2']))) else x['BsmtQual'], axis=1)
test['BsmtCond']=test[['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1',
                         'BsmtFinType2']].apply(lambda x: 'No Basement' if (x['BsmtQual']=='No Basement') and pd.isnull(x['BsmtCond']) and
                                                pd.isnull(x['BsmtExposure']) and pd.isnull(x['BsmtFinType1']) and 
                                                pd.isnull(x['BsmtFinType2']) else 'TA' if pd.isnull(x['BsmtCond']) and np.all((pd.notnull(x['BsmtQual'])))
                                                and np.all((pd.notnull(x['BsmtExposure']))) and np.all((pd.notnull(x['BsmtFinType1']))) and 
                                                np.all((pd.notnull(x['BsmtFinType2']))) else x['BsmtCond'], axis=1)
test['BsmtExposure']=test[['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1',
                         'BsmtFinType2']].apply(lambda x: 'No Basement' if (x['BsmtQual']=='No Basement') and (x['BsmtCond']=='No Basement') and
                                                pd.isnull(x['BsmtExposure']) and pd.isnull(x['BsmtFinType1']) and 
                                                pd.isnull(x['BsmtFinType2']) else 'No' if pd.isnull(x['BsmtExposure']) and np.all((pd.notnull(x['BsmtCond'])))
                                                and np.all((pd.notnull(x['BsmtQual']))) and np.all((pd.notnull(x['BsmtFinType1']))) and 
                                                np.all((pd.notnull(x['BsmtFinType2']))) else x['BsmtExposure'], axis=1)
test['BsmtFinType1']=test.BsmtFinType1.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
test['BsmtFinType2']=test.BsmtFinType2.apply(lambda x: 'No Basement' if pd.isnull(x) else x)
```

# Outlier Detection and Handling

#### Train Data

```python
column=[]
outlier_count=[]
non_outlier_count=[]
lower_limit=[]
upper_limit=[]
for c in train.select_dtypes([np.number]).columns:
    upper,lower=(train[c].mean()+train[c].std()*3),(train[c].mean()-train[c].std()*3)
    outlier=len([x for x in train[c] if x<lower or x>upper])
    non_outlier=len([x for x in train[c] if x>=lower and x<=upper])
    column.append(c)
    outlier_count.append(outlier)
    non_outlier_count.append(non_outlier)
    lower_limit.append(lower)
    upper_limit.append(upper)
outlier_dict=dict()
outlier_dict['column']=column
outlier_dict['outlier_count']=outlier_count
outlier_dict['non_outlier_count']=non_outlier_count
outlier_dict['lower_limit']=lower_limit
outlier_dict['upper_limit']=upper_limit

outlier=pd.DataFrame(outlier_dict)
outlier
```

```python
def outlier_treatment(df,columns):
    for col in columns:
        lower_limit=(outlier[(outlier.column==col)]['lower_limit'].values)[0]
        upper_limit=(outlier[(outlier.column==col)]['upper_limit'].values)[0]
        df[col]=np.where(df[col]>upper_limit,upper_limit,df[col])
        df[col]=np.where(df[col]<lower_limit,lower_limit,df[col])

cols=[x for x in outlier.loc[(outlier.outlier_count>0) & (outlier.upper_limit>20),'column']]
outlier_treatment(train,cols)
```

```python
ct=ColumnTransformer([('ohe',OneHotEncoder(sparse=False, handle_unknown='ignore'),
                       ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
       'MasVnrType', 'Foundation', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
       'Electrical', 'Functional', 'GarageType', 'GarageFinish', 'PavedDrive',
       'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'])],remainder='passthrough')
```

```python
models=[LinearRegression(), Lasso(random_state=1),Ridge(random_state=1),DecisionTreeRegressor(random_state=1)]
scales=[RobustScaler(),StandardScaler(),Normalizer(),MinMaxScaler()]
result=[]
for model in models:
    final=[]
    for scale in scales:
        pipe=Pipeline([('ct',ct),('scale',scale),('model',model)])
        score=cross_val_score(estimator=pipe,X=x_train,y=np.array(y_train).ravel(),cv=5,n_jobs=-1,scoring='r2').mean()
        final.append(round(score*100,2))
    result.append(final)
base_result=pd.DataFrame(result,index=['Linear Regression','Lasso','Ridge', 'Decision Tree Regression'],
             columns=['Robust Scaler','Standard Scaler','Normalizer','Min Max Scaler'])
base_result
```
